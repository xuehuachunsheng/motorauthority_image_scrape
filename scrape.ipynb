```python
# 爬取www.motorauthority.com网站上的图片
import bs4 as bs
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By

from urllib.request import Request, urlopen
import pandas as pd
import os
import re
import sys
import time
import json

# params
base_url = "https://www.motorauthority.com/pictures/page-{}"
headers = {'User-Agent': 'Mozilla/5.0 (Linux; Android 5.1.1; SM-G928X Build/LMY47X) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.83 Mobile Safari/537.36'}
page_ranges = [1, 500]
MAX_TIME_OUT = 3600

prefs = {"profile.managed_default_content_settings.images": 2,'permissions.default.stylesheet':2}
options = Options()
# options.add_argument('--headless') # 无浏览器模式
options.add_argument("user-agent=Mozilla/5.0 (Linux; Android 5.1.1; SM-G928X Build/LMY47X) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.83 Mobile Safari/537.36")
options.add_experimental_option("prefs", prefs)
driver = webdriver.Chrome('/Users/wuyanxue/Documents/安装包/chromedriver', options=options)
```

```python

img_text_url_mapping = {"page_num": 0}
total_fm_imgs = 0
for p in range(page_ranges[0], page_ranges[1]):
    page_url = base_url.format(p)
    driver.get(page_url)
    WebDriverWait(driver,MAX_TIME_OUT,0.5).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"img-wrap")))
    fengmian_imgs = driver.find_elements_by_class_name("img-wrap") # 获取封面图
    for fm_im in fengmian_imgs:
        print(f'\r{total_fm_imgs}/th image done..., current page: {p}', end='')
        fm_im.click()
        WebDriverWait(driver,MAX_TIME_OUT,0.5).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"image-wrapper")))
        soup = bs.BeautifulSoup(driver.page_source, 'lxml')
        # 获取图片
        divs = soup.findAll("div", {"class": "image-wrapper"})
        for d in divs:
            im_content = d.img
            im_url = im_content['src']
            im_text = im_content['alt'].strip()
            if im_text.strip() not in img_text_url_mapping:
                img_text_url_mapping[im_text] = []
            img_text_url_mapping[im_text].append(im_url)
        WebDriverWait(driver,MAX_TIME_OUT,0.5).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"hgm-overlay-close")))
        driver.find_element_by_class_name('hgm-overlay-close').click()
        total_fm_imgs += 1
    img_text_url_mapping['page_num'] = p
    
    if p % 10 == 0: # 每隔10页保存1次
        print(f'\n Storing... Current Page: {p}')
        with open(f'./data/meta-max-page{p}.json', 'w') as f:
            json.dump(img_text_url_mapping, f)
```
